```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv(r'C:\Users\79375\Desktop\DS\Mall_Customers\Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

```python
dataset.head()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1], s=100, c='red',label='Claster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1], s=100, c='blue',label='Claster 2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1], s=100, c='green',label='Claster 3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1], s=100, c='cyan',label='Claster 4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1], s=100, c='magenta',label='Claster 5')

plt.title('Clusters of Castomers')
plt.xlabel('Annusal Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
